## In this project we are using Tensorflow to build a Nueral Network

```python
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
```

## data generation

```python
observations=1000
xs=np.random.uniform(low=-10,high=10,size=(observations,1)) #draws a random value from the interval where each number has an equal chance to be selected size=1000*1
zs=np.random.uniform(-10,10,(observations,1)) #size=1000*1
#size=n*k
generated_inputs=np.column_stack((xs,zs)) #inputs matix it satck the obseravtions into matrix

noise=np.random.uniform(-1,1,(observations,1))
generated_targets= 2*xs-3*zs+5+ noise
np.savez('TF_intro',inputs=generated_inputs,targets=generated_targets) # its saves the n-d arrays in .npz files which is required for tensorflow
#tf_intro is the file name
```

## Loading the file

```python
# solving using the tensorflow
training_data= np.load('TF_intro.npz')
```

## Applying Xavier Glorot Initialization

```python
inputs_size=2
output_size=1
model=tf.keras.initializers.GlorotUniform
# intensorflow we actually build the model
model=tf.keras.Sequential([
    tf.keras.layers.Dense(output_size) # takes inputs and calculates the dot product of the inputs and weights and adds the bias
                    

]) # it allows the model to be laid down in form of stack layers
 # configures the model for training that is the optimization algo which here is stochastic gradient descent
#it takes two arguements 1st one is the optimization algo and the second one is the the loss functiomn which here is the l2 norm loss
model.compile(optimizer='sgd',loss='mean_squared_error')
model.fit(training_data['inputs'],training_data['targets'],epochs=100,verbose=2) #we fit what we need to train here the epoch is the no of iterations

# this is basically the loop iteration which we had done in the previous lecture
```

## Extract the outputs

```python
model.predict_on_batch(training_data['inputs']).round(1)
```

## Plotting the results

```python
# plotting the graph
plt.plot(np.squeeze(model.predict_on_batch(training_data['inputs'])),np.squeeze(training_data['targets']))
plt.xlabel('outputs')
plt.ylabel('targets')
plt.show()
```
